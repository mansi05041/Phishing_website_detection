# LOGISTIC REGRESSON

```python
# importing variables from pre processing
%store -r X_train
%store -r X_test
%store -r y_train
%store -r y_test
%store -r label
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
# importing libraries
import matplotlib.pyplot as plt
```

# TRAINING THE MODEL

```python
lr = LogisticRegression(max_iter=10000)
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

# EVALUATION OF MODEL

```python
# calculate f1 score
lr_f1 = f1_score(y_test, y_pred)
print("F1-score of logistic regression: ",lr_f1)
```

```python
# calculate accuraccy
lr_accuracy = accuracy_score(y_test,y_pred)
print("Accuracy of logistic regression : ",lr_accuracy)
```

```python
# calculate precision
lr_precision = precision_score(y_test,y_pred)
print("Precision of logistic regression: ",lr_precision)
```

```python
# calculate recall score
lr_recall = recall_score(y_test,y_pred)
print("Recall of logistic regression: ",lr_recall)
```

```python
# Evaluate the roc

# compute the ROC curve and AUC for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(len(label)):
    fpr[i], tpr[i], _ = roc_curve(y_test,y_pred)
    roc_auc[i]=auc(fpr[i],tpr[i])

# plot the Roc curve for each class
plt.figure(figsize=(8,6))
for i in range(len(label)):
    plt.plot(fpr[i],tpr[i], label='ROC curve (AUC = %0.2f)' % roc_auc[i])
plt.plot([0,1],[0,1],color='navy',linestyle='--')
plt.xlabel('False Positive rate')
plt.ylabel('True Positive rate')
plt.title('ROC')
plt.legend(loc="lower right")
plt.show()
```

```python
# confusion matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
cm_display = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=label)
cm_display.plot()
plt.title('Confusion Matrix')
plt.show()
```
